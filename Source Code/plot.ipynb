Plot Data Splitting Technique

```python
import pandas as pd
import matplotlib.pyplot as plt

# Paths to the CSV files
file_paths = [
    r'C:\Users\dhani\Documents\Tugas Akhir and Stuff\Val Code\training_results_with_34_0.csv',  # Example: file for Manual Split
    r'C:\Users\dhani\Documents\Tugas Akhir and Stuff\Val Code\training_results_with_34_0_random.csv',    # Example: file for Random Split
]

# Corresponding labels for each CSV file
csv_name = ['Manual Split', 'Random Split']

# Initialize variables to track the highest global accuracy and F1 score
highest_acc = 0
highest_acc_epoch = 0
highest_acc_name = None

highest_f1 = 0
highest_f1_epoch = 0
highest_f1_name = None

# Validation Accuracy Plot
plt.figure(figsize=(12, 4))
for i, file_path in enumerate(file_paths):
    # Read the data
    data = pd.read_csv(file_path)

    # Extract the epochs and validation accuracy
    epochs = data['Epoch']
    val_acc = data['Val Accuracy']

    # Plot validation accuracy
    plt.plot(epochs, val_acc, label=f'{csv_name[i]}')

    # Find the highest accuracy for this split
    max_acc = val_acc.max()
    max_acc_epoch = epochs[val_acc.idxmax()]
    plt.scatter(max_acc_epoch, max_acc, color='red', s=50,
                label=f'{csv_name[i]} Top Acc: {max_acc:.2f}% (Epoch {max_acc_epoch})')

    # Track the highest accuracy globally
    if max_acc > highest_acc:
        highest_acc = max_acc
        highest_acc_epoch = max_acc_epoch
        highest_acc_name = csv_name[i]

plt.title('Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy (%)')
plt.grid(True)
plt.legend()
plt.show()

# Validation F1 Score Plot
plt.figure(figsize=(12, 4))
for i, file_path in enumerate(file_paths):
    # Read the data
    data = pd.read_csv(file_path)

    # Extract the epochs and validation F1 scores
    epochs = data['Epoch']
    val_f1_scores = data['Val F1 Score']

    # Plot validation F1 score
    plt.plot(epochs, val_f1_scores, label=f'{csv_name[i]}')

    # Find the highest F1 score for this split
    max_f1 = val_f1_scores.max()
    max_f1_epoch = epochs[val_f1_scores.idxmax()]
    plt.scatter(max_f1_epoch, max_f1, color='red', s=50,
                label=f'{csv_name[i]} Top F1: {max_f1:.4f} (Epoch {max_f1_epoch})')

    # Track the highest F1 score globally
    if max_f1 > highest_f1:
        highest_f1 = max_f1
        highest_f1_epoch = max_f1_epoch
        highest_f1_name = csv_name[i]

plt.title('Validation F1 Score')
plt.xlabel('Epoch')
plt.ylabel('Validation F1 Score')
plt.grid(True)
plt.legend()
plt.show()
```
